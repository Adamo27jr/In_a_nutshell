```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt

# Import des donnes
ozone = pd.read_csv('ozone_1.csv', sep=';', decimal=',')

# 1. Matrice des donnes manquantes
plt.figure(figsize=(15, 10))
msno.matrix(ozone, color=(0.8, 0.1, 0.1), fontsize=9)
plt.title('Matrixplot des donnes manquantes - ozone_4.csv', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

# 2. Heatmap de corrlation des manquants
plt.figure(figsize=(12, 8))
msno.heatmap(ozone, cmap='Reds', fontsize=10)
plt.title('Corrlation entre les patterns de donnes manquantes', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

numeric_vars = ['maxO3', 'T9', 'T12', 'T15', 'Ne9', 'Ne12', 'Ne15', 'Vx9', 'Vx12', 'Vx15', 'maxO3v']

# 3. Analyse par variable catgorielle
categorical_vars = ['vent', 'pluie']
for cat_var in categorical_vars:
    if cat_var in ozone.columns:
        plt.figure(figsize=(12, 5))
        
        # Proportion de manquants par catgorie
        missing_by_cat = ozone.groupby(cat_var).apply(lambda x: x.isnull().mean())
        missing_by_cat[numeric_vars].plot(kind='bar', figsize=(12, 6))
        plt.title(f'Proportion de donnes manquantes par {cat_var}')
        plt.ylabel('Proportion de manquants')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, chi2_contingency

# Charger les donnes
ozone = pd.read_csv('ozone_2.csv', sep=';', decimal=',')

# Variables de nbulosit
nebulosite_vars = ['Ne9', 'Ne12', 'Ne15']

print("=== ANALYSE LIEN NBULOSIT - DONNES MANQUANTES ===")

for ne_var in nebulosite_vars:
    print(f"\n--- Analyse pour {ne_var} ---")
    
    # Vrifier si la variable existe et a des donnes
    if ne_var in ozone.columns and ozone[ne_var].notna().any():
        
        # 1. Distribution de la nbulosit quand d'autres variables sont manquantes
        for target_var in ['T9', 'T12', 'T15']:
            if target_var in ozone.columns:
                # Crer des groupes : nbulosit quand la variable cible est manquante vs prsente
                mask_missing = ozone[target_var].isnull()
                nebulosite_present = ozone.loc[~mask_missing, ne_var].dropna()
                nebulosite_missing = ozone.loc[mask_missing, ne_var].dropna()
                
                if len(nebulosite_present) > 5 and len(nebulosite_missing) > 5:
                    # Test t pour comparer les moyennes
                    t_stat, p_value = ttest_ind(nebulosite_present, nebulosite_missing, 
                                               nan_policy='omit', equal_var=False)
                    
                    print(f"Quand {target_var} est manquant vs prsent:")
                    print(f"  Nbulosit moyenne (prsent): {nebulosite_present.mean():.2f}")
                    print(f"  Nbulosit moyenne (manquant): {nebulosite_missing.mean():.2f}")
                    print(f"  p-value: {p_value:.4f}")
                    if p_value < 0.05:
                        print(f"   Diffrence SIGNIFICATIVE (niveau = {p_value:.4f})")
                    else:
                        print(f"   Pas de diffrence significative")
                    print()
        
        # 2. Analyse par catgories de nbulosit
        plt.figure(figsize=(10, 6))
        
        # Catgoriser la nbulosit (faible, moyenne, forte)
        ozone[f'{ne_var}_cat'] = pd.cut(ozone[ne_var], 
                                       bins=[-1, 2, 5, 10], 
                                       labels=['Faible', 'Moyenne', 'Forte'],
                                       include_lowest=True)
        
        # Calculer le taux de manquants pour chaque variable par catgorie de nbulosit
        missing_rates = ozone.groupby(f'{ne_var}_cat').apply(
            lambda x: x[['T9', 'T12', 'T15', 'maxO3']].isnull().mean()
        )
        
        missing_rates.plot(kind='bar', figsize=(12, 6))
        plt.title(f'Taux de donnes manquantes par catgorie de {ne_var}')
        plt.ylabel('Proportion de donnes manquantes')
        plt.xlabel('Catgorie de nbulosit')
        plt.legend(title='Variables')
        plt.tight_layout()
        plt.show()
        
        # 3. Test du chi pour chaque variable
        for target_var in ['T9', 'T12', 'T15']:
            if target_var in ozone.columns:
                # Table de contingence: catgorie nbulosit vs prsence donne cible
                contingency_table = pd.crosstab(
                    ozone[f'{ne_var}_cat'], 
                    ozone[target_var].isnull(),
                    margins=True
                )
                
                chi2, p_value, dof, expected = chi2_contingency(contingency_table.iloc[:-1, :-1])
                print(f"Test chi {ne_var} vs manquants {target_var}: p-value = {p_value:.4f}")
```
