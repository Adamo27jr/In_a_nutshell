```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from tqdm import tqdm
import os

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class SinusoidalPositionEmbeddings(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.dim = dim

    def forward(self, time):
        device = time.device
        half_dim = self.dim // 2
        embeddings = np.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=device) * -embeddings)
        embeddings = time[:, None] * embeddings[None, :]
        embeddings = torch.cat((embeddings.sin(), embeddings.cos()), dim=-1)
        return embeddings

class SimpleBlock(nn.Module):
    def __init__(self, in_channels, out_channels, time_emb_dim):
        super().__init__()
        
        self.conv1 = nn.Conv2d(in_channels, out_channels, 3, padding=1)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1)
        self.time_mlp = nn.Linear(time_emb_dim, out_channels)
        
    def forward(self, x, t_emb):
        h = F.silu(self.conv1(x))
        
        time_emb = self.time_mlp(t_emb)
        time_emb = time_emb[(..., ) + (None, ) * 2]
        h = h + time_emb
        
        h = F.silu(self.conv2(h))
        return h

class SimpleMNistUNet(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, time_emb_dim=64):
        super().__init__()
        
        self.time_mlp = nn.Sequential(
            SinusoidalPositionEmbeddings(time_emb_dim),
            nn.Linear(time_emb_dim, time_emb_dim),
            nn.SiLU(),
            nn.Linear(time_emb_dim, time_emb_dim)
        )

        self.enc1 = SimpleBlock(in_channels, 32, time_emb_dim)
        self.enc2 = SimpleBlock(32, 64, time_emb_dim)
        self.enc3 = SimpleBlock(64, 128, time_emb_dim)

        self.bottleneck = SimpleBlock(128, 256, time_emb_dim)

        self.up1 = nn.ConvTranspose2d(256, 128, 4, stride=2, padding=1)
        self.dec1 = SimpleBlock(256, 128, time_emb_dim)  
        
        self.up2 = nn.ConvTranspose2d(128, 64, 4, stride=2, padding=1)
        self.dec2 = SimpleBlock(128, 64, time_emb_dim)  
        
        self.up3 = nn.ConvTranspose2d(64, 32, 4, stride=2, padding=1)
        self.dec3 = SimpleBlock(64, 32, time_emb_dim)    

        self.final_conv = nn.Conv2d(32, out_channels, 3, padding=1)
        
    def forward(self, x, t):
        t_emb = self.time_mlp(t)

        x1 = self.enc1(x, t_emb)                    
        x2 = self.enc2(F.max_pool2d(x1, 2), t_emb)  
        x3 = self.enc3(F.max_pool2d(x2, 2), t_emb) 
        
        x_bottleneck = self.bottleneck(F.max_pool2d(x3, 2), t_emb) 

        x = self.up1(x_bottleneck) 

        if x.shape[2:] != x3.shape[2:]:
            x = F.interpolate(x, size=x3.shape[2:], mode='nearest')
        
        x = torch.cat([x, x3], dim=1) 
        x = self.dec1(x, t_emb)        
        
        x = self.up2(x) 
        if x.shape[2:] != x2.shape[2:]:
            x = F.interpolate(x, size=x2.shape[2:], mode='nearest')
            
        x = torch.cat([x, x2], dim=1)
        x = self.dec2(x, t_emb)    
        
        x = self.up3(x)  
        if x.shape[2:] != x1.shape[2:]:
            x = F.interpolate(x, size=x1.shape[2:], mode='nearest')
            
        x = torch.cat([x, x1], dim=1)  
        x = self.dec3(x, t_emb)      
        
        return self.final_conv(x)

class DDPM:
    def __init__(self, model, timesteps=1000, beta_start=1e-4, beta_end=0.02, device="cuda"):
        self.model = model
        self.timesteps = timesteps
        self.device = device

        self.betas = torch.linspace(beta_start, beta_end, timesteps, device=device)
        self.alphas = 1. - self.betas
        self.alpha_bars = torch.cumprod(self.alphas, dim=0)
        
        print(f"Beta schedule: {beta_start} -> {beta_end}")
        print(f"Alpha bars: {self.alpha_bars[0]:.3f} -> {self.alpha_bars[-1]:.3f}")
        print(f"Nombre de timesteps: {timesteps}")
        
    def forward_process(self, x0, t, noise=None):
        if noise is None:
            noise = torch.randn_like(x0)

        t = torch.clamp(t, 0, self.timesteps - 1)
        
        alpha_bar_t = self.alpha_bars[t].view(-1, 1, 1, 1)
        x_t = torch.sqrt(alpha_bar_t) * x0 + torch.sqrt(1 - alpha_bar_t) * noise
        
        return x_t
    
    def reverse_process(self, x, t):
        t = torch.clamp(t, 0, self.timesteps - 1)
        
        pred_noise = self.model(x, t)
        
        alpha_t = self.alphas[t].view(-1, 1, 1, 1)
        alpha_bar_t = self.alpha_bars[t].view(-1, 1, 1, 1)
        beta_t = self.betas[t].view(-1, 1, 1, 1)
        
        if t[0] > 0:
            mean = (x - beta_t * pred_noise / torch.sqrt(1 - alpha_bar_t)) / torch.sqrt(alpha_t)

            noise = torch.randn_like(x)
            std = torch.sqrt(beta_t)
            x_prev = mean + std * noise
        else:
            x_prev = (x - torch.sqrt(1 - alpha_bar_t) * pred_noise) / torch.sqrt(alpha_bar_t)
        
        return x_prev
    
    def train_step(self, x0, optimizer):
        self.model.train()

        t = torch.randint(0, self.timesteps, (x0.shape[0],), device=self.device)

        noise = torch.randn_like(x0)

        x_t = self.forward_process(x0, t, noise)

        pred_noise = self.model(x_t, t)

        loss = F.mse_loss(pred_noise, noise)
        
        loss.backward()
        optimizer.step()
        
        return loss.item()
    
    def sample(self, num_samples=16, image_size=(28, 28)):
        self.model.eval()
        
        with torch.no_grad():
            x = torch.randn(num_samples, 1, image_size[0], image_size[1], device=self.device)

            for i in tqdm(reversed(range(self.timesteps)), desc="Génération"):
                t = torch.full((num_samples,), i, device=self.device, dtype=torch.long)
                x = self.reverse_process(x, t)

            x = torch.clamp(x, -1., 1.)
            x = (x + 1.) / 2.
            
        return x

def visualize_diffusion_process(ddpm, dataloader):
    sample_batch, _ = next(iter(dataloader))
    sample_batch = sample_batch.to(device)

    fig, axes = plt.subplots(2, 6, figsize=(15, 5))

    axes[0, 0].imshow(sample_batch[0, 0].cpu().numpy(), cmap='gray')
    axes[0, 0].set_title('Original (t=0)')
    axes[0, 0].axis('off')
    
    max_t = ddpm.timesteps - 1 
    timesteps = [0] + [int(max_t * i / 5) for i in range(1, 6)]
    
    print(f"Timesteps utilisés pour la visualisation: {timesteps}")
    
    for i, t in enumerate(timesteps[1:], 1):
        t_batch = torch.full((1,), t, device=device)
        x_noisy = ddpm.forward_process(sample_batch[:1], t_batch)
        axes[0, i].imshow(x_noisy[0, 0].cpu().numpy(), cmap='gray')
        axes[0, i].set_title(f't = {t}')
        axes[0, i].axis('off')
    
    axes[1, 0].text(0.5, 0.5, 'Processus de bruitage\n(Forward Process)', 
                   ha='center', va='center', transform=axes[1, 0].transAxes, fontsize=12)
    axes[1, 0].axis('off')
    
    for i in range(1, 6):
        axes[1, i].text(0.5, 0.5, f'Bruit: {timesteps[i]}/{max_t}', 
                       ha='center', va='center', transform=axes[1, i].transAxes, fontsize=10)
        axes[1, i].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return timesteps

def load_mnist(batch_size=64):
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    
    train_dataset = torchvision.datasets.MNIST(
        root='./data', 
        train=True, 
        download=True, 
        transform=transform
    )
    
    train_loader = DataLoader(
        train_dataset, 
        batch_size=batch_size, 
        shuffle=True,
        num_workers=2
    )
    
    return train_loader

def train_ddpm_mnist():
    train_loader = load_mnist(batch_size=64)
    print("MNIST dataset loaded")

    model = SimpleMNistUNet(in_channels=1, out_channels=1, time_emb_dim=64).to(device)

    ddpm = DDPM(model, timesteps=200, beta_start=1e-4, beta_end=0.02, device=device)

    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    
    print("Visualisation du processus de diffusion...")
    timesteps_used = visualize_diffusion_process(ddpm, train_loader)

    losses = []
    num_epochs = 100
    
    print("Début de l'entraînement...")
    
    for epoch in range(num_epochs):
        epoch_loss = 0
        num_batches = 0
        
        for batch_idx, (data, _) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")):
            data = data.to(device)
            
            loss = ddpm.train_step(data, optimizer)
            epoch_loss += loss
            num_batches += 1
        
        avg_loss = epoch_loss / num_batches
        losses.append(avg_loss)
        print(f"Epoch {epoch+1}, Loss: {avg_loss:.4f}")
        
        if (epoch + 1) % 2 == 0:
            print("Génération d'échantillons...")
            samples = ddpm.sample(num_samples=8, image_size=(28, 28))
            
            fig, axes = plt.subplots(2, 4, figsize=(12, 6))
            for i, ax in enumerate(axes.flat):
                if i < len(samples):
                    ax.imshow(samples[i, 0].cpu().numpy(), cmap='gray')
                    ax.axis('off')
            plt.suptitle(f'Époch {epoch+1} - Loss: {avg_loss:.4f}')
            plt.tight_layout()
            plt.show()
    
    plt.figure(figsize=(10, 5))
    plt.plot(losses)
    plt.title('Loss pendant l\'entraînement')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()
    
    return ddpm

def demonstrate_final_results(ddpm):
    
    print("Génération de chiffres MNIST...")
    
    samples = ddpm.sample(num_samples=24, image_size=(28, 28))
    
    fig, axes = plt.subplots(4, 6, figsize=(15, 10))
    for i, ax in enumerate(axes.flat):
        if i < len(samples):
            ax.imshow(samples[i, 0].cpu().numpy(), cmap='gray')
            ax.axis('off')
    
    plt.suptitle('Chiffres MNIST générés par le modèle de diffusion', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    print("Visualisation du processus de génération...")
    
    ddpm.model.eval()
    with torch.no_grad():
        x = torch.randn(1, 1, 28, 28, device=device)

        total_steps = ddpm.timesteps
        steps_to_show = [0] + [int((total_steps - 1) * p) for p in [0.2, 0.4, 0.6, 0.8, 0.95]]
        
        intermediate = []
        
        for i in tqdm(reversed(range(total_steps)), desc="Processus de génération"):
            t = torch.full((1,), i, device=device, dtype=torch.long)
            x = ddpm.reverse_process(x, t)
            
            if i in steps_to_show:
                img = x.clone()
                img = torch.clamp(img, -1., 1.)
                img = (img + 1.) / 2.
                intermediate.append((total_steps - i, img.cpu()))

        fig, axes = plt.subplots(2, 3, figsize=(12, 8))
        for idx, (step, img) in enumerate(intermediate):
            ax = axes[idx // 3, idx % 3]
            ax.imshow(img[0, 0].numpy(), cmap='gray')
            ax.set_title(f'Step {step}/{total_steps}')
            ax.axis('off')
        
        plt.suptitle('Évolution: du bruit vers le chiffre', fontsize=16)
        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    print("=== Modèle de Diffusion pour MNIST ===\n")
    
    try:
        print("1. Entraînement du modèle...")
        trained_ddpm = train_ddpm_mnist()
        
        print("\n2. Démonstration des résultats...")
        demonstrate_final_results(trained_ddpm)
        
        print("\n3. Génération de plus d'échantillons...")
        final_samples = trained_ddpm.sample(num_samples=36, image_size=(28, 28))
        
        fig, axes = plt.subplots(6, 6, figsize=(12, 12))
        for i, ax in enumerate(axes.flat):
            ax.imshow(final_samples[i, 0].cpu().numpy(), cmap='gray')
            ax.axis('off')
        
        plt.suptitle('36 chiffres MNIST générés par le modèle DDPM', fontsize=16)
        plt.tight_layout()
        plt.show()
        
        print("\nDémonstration terminée avec succès!")
        print("Le modèle a appris à générer des chiffres MNIST à partir de bruit!")
        
    except Exception as e:
        print(f"\nErreur: {e}")
        import traceback
        traceback.print_exc()
```
