## Exercice 0 : Commandes de base Docker
Voici qques commandes essentielles à tester soit dans un terminal (de préférence) soit dans le notebook avec les `!` :
- `docker --version` → Vérifie la version
- `docker ps` → Liste les conteneurs actifs
- `docker images` → Liste les images disponibles
- `docker run` → Lance un conteneur à partir d’une image
- `docker stop <id>` → Arrête un conteneur
- `docker rm <id>` → Supprime un conteneur
- `docker rmi <image>` → Supprime une image

## Exercice Bonus : Nettoyer Docker
Pour éviter d'accumuler des conteneurs/images inutilisés :
- `docker ps -a` → Voir tous les conteneurs
- `docker rm <id>` → Supprimer un conteneur arrêté
- `docker rmi <image>` → Supprimer une image
- `docker system prune` → Nettoyer tout (⚠️ supprime ce qui est inutile).
