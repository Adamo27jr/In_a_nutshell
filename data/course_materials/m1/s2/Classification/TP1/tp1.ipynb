```r
# Définition des points
point1 <- c(-2, -1)
point2 <- c(2, 3)
origin <- c(0, 0)

# Matrices de covariance
cov_A <- matrix(c(1, 0, 0, 1), 2, 2)
cov_B <- matrix(c(1, 0, 0, 5), 2, 2)
cov_C <- matrix(c(1, 2, 2, 5), 2, 2)

# Distance euclidienne
euclidean_distance <- function(p1, p2) {
  sqrt(sum((p1 - p2)^2))
}

# Distances entre les points
d_euc <- euclidean_distance(point1, point2)
d_euc_origin <- euclidean_distance(point1, origin)

# Distance de Mahalanobis
mahalanobis_distance <- function(p1, p2, cov_matrix) {
  diff <- p1 - p2
  sqrt(t(diff) %*% solve(cov_matrix) %*% diff)
}

# Distances de Mahalanobis
d_mah_A <- mahalanobis_distance(point1, point2, cov_A)
d_mah_origin <- mahalanobis_distance(point1, origin, cov_A)

# Afficher les résultats
print(paste("Distance Euclidienne entre point1 et point2 : ", d_euc))
print(paste("Distance Euclidienne entre point1 et l'origine : ", d_euc_origin))
print(paste("Distance de Mahalanobis entre point1 et point2 (Cov_A) : ", d_mah_A))
print(paste("Distance de Mahalanobis entre point1 et l'origine (Cov_A) : ", d_mah_origin))
```
