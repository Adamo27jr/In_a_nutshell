```python
import numpy as np
import matplotlib.pyplot as plt

def alpha_critique(p):
    if p < 4/5 :
        return (np.sqrt(4) / 5) * np.sqrt(1 / (p * (1 - p)))
    else :
        return 1

p_values = np.linspace(0.500, 1.000, 21)

# Calcul des valeurs alpha correspondantes
alpha_critiques = [alpha_critique(p) for p in p_values]

# Tracé de la courbe avec le style spécifié
plt.figure(figsize=(10, 6))
plt.plot(p_values, alpha_critiques, 'o-', label='α critique')
plt.xlabel('Probabilité p de mouvement vers la droite')
plt.ylabel('α critique')
plt.title('Valeur critique de α en fonction de p')
plt.grid(True)
plt.legend()
plt.show()
```
