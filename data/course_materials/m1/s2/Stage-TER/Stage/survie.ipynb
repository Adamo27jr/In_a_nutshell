```python
import numpy as np
import matplotlib.pyplot as plt

# Définition de la fonction
def f(i, A, B, C):
    return np.exp(-A * i - B * ( (C**i - 1) / (C - 1) ) )

# Paramètres (à ajuster selon vos besoins)
A = 0.001
B = 0.001
C = 1.01

# Création d'un array de valeurs pour i
i_values = np.linspace(0, 500, 500)  

# Calcul des valeurs de la fonction
y_values = f(i_values, A, B, C)

# Tracé du graphe
plt.figure(figsize=(10, 6))
plt.plot(i_values, y_values, label=f'$f(i) = \exp(-{A}i - {B}\\frac{{{C}^i - 1}}{{{C} - 1}})$', color='blue')
plt.xlabel('i', fontsize=12)
plt.ylabel('f(i)', fontsize=12)
plt.title('Graphe de $f(i) = \exp(-Ai - B \\frac{C^{i}-1}{C-1})$', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètre alpha (à modifier selon vos besoins)
alpha = 0.96  # Exemple : 0 < alpha < 1 pour une décroissance exponentielle

i_values = np.linspace(0, 150, 150)  

# Calcul de alpha^i
alpha_powers = alpha ** i_values

# Tracé du graphe
plt.figure(figsize=(10, 6))
plt.plot(i_values, alpha_powers, label=f'$\\alpha^i$ avec $\\alpha = {alpha}$', color='blue')
plt.xlabel('i', fontsize=12)
plt.ylabel('$\\alpha^i$', fontsize=12)
plt.title('Courbe de $\\alpha^i$ en fonction de $i$', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```
